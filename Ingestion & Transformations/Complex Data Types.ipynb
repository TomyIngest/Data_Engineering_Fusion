##### Reading and Exploring Data
- `Reading JSON is like reading any other file`
- `JSON string is nested text structure where we can read nested structures using (:)`
- `Another method is to parse json String using from_json function`
- `from_json fucntion requires schema which will be used for parsing, we can get it by using fuction       schema_of_json`

```python
%sql
-- selecting nested structures from json string using (:) method
SELECT
    -- headers,
    value,
    value:eventId,
    value:data,
    value:data:orderID
FROM read_files(
    '/Volumes/data_fusion/source_data/v02_json/Kafka_Stream/',
    format => 'json'
)LIMIT 1
```

```python
%sql
-- Reading schema from json String
SELECT
    schema_of_json(value) as schema_JSON
FROM read_files(
    '/Volumes/data_fusion/source_data/v02_json/Kafka_Stream/',
    format => 'json'
) LIMIT 1
```

```python
%sql
-- parsing JSON string type into STRUCT type which is much faster
SELECT
    headers,
    array_size(headers),
    explode(headers) as exploded,
    exploded.key,
    exploded.value,
    value,
    from_json(value,'STRUCT<data: STRUCT<amount: DOUBLE, currency: STRING, method: STRING, orderId: STRING, processor: STRUCT<code: STRING, message: STRING, name: STRING>>, eventId: STRING, eventType: STRING, eventVersion: STRING, occurredAt: BIGINT, source: STRING, traceId: STRING>'
    ) as Json_parsed
FROM read_files(
    '/Volumes/data_fusion/source_data/v02_json/Kafka_Stream/',
    format => 'json'
)LIMIT 4
```

##### Python

```python
import pyspark.sql.functions as sf
```

```python
json_string = spark.read.format("json").load('/Volumes/data_fusion/source_data/v02_json/Kafka_Stream/') \
            .select("value").collect()[0][0]

json_schema = sf.schema_of_json(json_string)

df = spark.read.format("json").load('/Volumes/data_fusion/source_data/v02_json/Kafka_Stream/')
df = df.select(
        "headers",
        sf.size("headers"),
        sf.explode("headers").alias("exploded"),
        sf.col("exploded.key"),
        sf.col("exploded.value").alias("exploded_value"),
        sf.col("value").alias("Json_String")
        ) \
        .withColumn("JSON_Parsed", sf.from_json(sf.col("Json_String"),json_schema))

df.limit(4).display()
```
