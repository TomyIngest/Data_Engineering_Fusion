##### Reading and Exploring Data
- `Reading JSON is like reading any other file`
- `JSON string is nested text structure where we can read nested structures using (:)`
- `Another method is to parse json String using from_json function`
- `from_json fucntion requires schema which will be used for parsing, we can get it by using fuction       schema_of_json`

```python
%sql
-- selecting nested structures from json string using (:) method
SELECT
    -- headers,
    value,
    value:eventId,
    value:data,
    value:data:orderID
FROM read_files(
    '/Volumes/data_fusion/source_data/v02_json/Kafka_Stream/',
    format => 'json'
)LIMIT 1
```

```python
%sql
-- Reading schema from json String
SELECT
    schema_of_json(value) as schema_JSON
FROM read_files(
    '/Volumes/data_fusion/source_data/v02_json/Kafka_Stream/',
    format => 'json'
) LIMIT 1
```

```python
%sql
-- parsing JSON string type into STRUCT type which is much faster
SELECT
    value,
    from_json(value,'STRUCT<data: STRUCT<amount: DOUBLE, currency: STRING, method: STRING, orderId: STRING, processor: STRUCT<code: STRING, message: STRING, name: STRING>>, eventId: STRING, eventType: STRING, eventVersion: STRING, occurredAt: BIGINT, source: STRING, traceId: STRING>'
    ) as Json_parsed
FROM read_files(
    '/Volumes/data_fusion/source_data/v02_json/Kafka_Stream/',
    format => 'json'
)LIMIT 1
```
